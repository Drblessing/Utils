```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
df = pd.DataFrame(columns = ['rank','price','id','last'])
for i in range(20):
    start_rank = 500*i+1
    end_rank = 500*i+500
    print('Querying rank', start_rank, 'to', end_rank)
    res = requests.get(
        f'https://howrare.is/boldbadgers/?rank_from={start_rank}&rank_to={end_rank}&for_sale=on&norm_traits=off'
    )
    soup = BeautifulSoup(res.content, 'html.parser')

    for badger in soup.find_all('div', class_= 'featured_image_desc'):
        rank = badger.findChild('div', class_="item_stat").findChild('span').text
        id_ = badger.find_all('div', class_="item_stat")[1].findChild('span').text.replace('#','')
        price = badger.find('a').text.strip().replace('buy ◎','').replace('Solanart','')
        try: 
            last = badger.find('div',class_='last_sale').text.strip().split('◎')[1]
        except:
            last = 0 
        df = pd.concat([df,pd.DataFrame(np.array([rank,price,id_,last]).reshape(1,4),columns = df.columns)])
df.index = df.id
```
