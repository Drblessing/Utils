```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
nft_collection = 'boldbadgers'
def get_collection(collection_name = 'boldbadgers', start_rank = 0, end_rank = 20):
    df = pd.DataFrame(columns = ['rank','price','id','last','marketplace'])
    for i in range(start_rank,end_rank):
        start_rank = 500*i+1
        end_rank = 500*i+500
        print('Querying rank', start_rank, 'to', end_rank)
        res = requests.get(
            f'https://howrare.is/{nft_collection}/?rank_from={start_rank}&rank_to={end_rank}&for_sale=on&norm_traits=off'
        )
        soup = BeautifulSoup(res.content, 'html.parser')
        for nft in soup.find_all('div', class_= 'featured_image_desc'):
            rank = nft.findChild('div', class_="item_stat").findChild('span').text
            id_ = nft.find_all('div', class_="item_stat")[1].findChild('span').text.replace('#','')
            price = nft.find('a').find('span').contents[0].text.replace('buy ◎','')
            marketplace = str(nft.find('a').find('span').contents[1])
            try: 
                last = nft.find('div',class_='last_sale').text.strip().split('◎')[1]
            except:
                last = 0 
            df = pd.concat([df,pd.DataFrame(np.array([rank,price,id_,last,marketplace]).reshape(1,5),columns = df.columns)])
    df.index = df.id
    return df
```

```python
df = get_collection()
```

```python
df_quantumtraders = get_collection('quantumtraders',0,18)
```
